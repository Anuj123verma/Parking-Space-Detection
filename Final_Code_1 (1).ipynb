```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.layers import Input,Dense,Conv2D, MaxPooling2D, UpSampling2D
from tensorflow.keras.layers import Flatten, Reshape,Dropout
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras import regularizers
from tensorflow.keras.utils import to_categorical
from keras.layers.normalization import BatchNormalization
from numba import jit,cuda
from imblearn.over_sampling import SMOTE
```

```
from skimage.transform import resize
train_images_pixels=[]
for i in range(0, len(train_files)):
  image=cv2.imread(train_files[i])
  image_resized= resize(image,(56,128,3))
  train_images_pixels.append(image_resized)
```

```
train_images_pixels=np.array(train_images_pixels)
#train_images_pixels = train_images_pixels.reshape((len(train_images_pixels),56,128,3))
```

```
X_train, X_test, y_train, y_test = train_test_split(
        train_images_pixels, train_correct_label, train_size=0.75, test_size=0.25, random_state=42, shuffle=True)
```

```
X_train_1, X_validate, y_train_1, y_validate = train_test_split(
        X_train, y_train, train_size=0.80, test_size=0.20, random_state=42, shuffle=True)
```

```
X_train_1=X_train_1.reshape((len(X_train_1),56,128,3))
X_validate=X_validate.reshape((len(X_validate),56,128,3))
X_test=X_test.reshape((len(X_test),56,128,3))
```

```
X_train1, X_test1, y_train1, y_test1 = train_test_split(
        train_images_pixels, train_correct_label, train_size=0.75, test_size=0.25, random_state=42, shuffle=True)
```

```
X_train_11, X_validate1, y_train_11, y_validate1 = train_test_split(
        X_train1, y_train1, train_size=0.80, test_size=0.20, random_state=42, shuffle=True)
```

```
from keras.regularizers import l2
from keras.optimizers import SGD
def build_model_final(train_images,train_labels1,X_valid_images,y_valid,batch_size=10,epochs=12,verbose=1):
    model=Sequential()
    model.add(Conv2D(32,(7,3),strides=(1, 1),activation='relu',kernel_regularizer=l2(0.000005),input_shape=(56,128,3)))
    #model.add(BatchNormalization())
    model.add(MaxPooling2D((2,2)))
    model.add(Conv2D(64,(6,4),strides=(1, 1),activation='relu',kernel_regularizer=l2(0.000005)))
    #model.add(BatchNormalization())
    model.add(MaxPooling2D((2,2)))
    #model.add(Dropout(0.25))
    #model.add(Conv2D(64,(3,3),strides=(1, 1),activation='relu'))
    #model.add(BatchNormalization())
    #model.add(Dropout(0.25))
    #model.add(Conv2D(128,(3,3),strides=(1, 1),activation='relu'))
    #model.add(BatchNormalization())
    #model.add(MaxPooling2D((2,2)))
    #model.add(Dropout(0.25))
    model.add(Flatten())
    model.add(Dense(50,activation='relu',kernel_regularizer=l2(0.000005)))
    #model.add(BatchNormalization())
    #model.add(Dropout(0.5))
    model.add(Dense(2,activation='relu',kernel_regularizer=l2(0.000005)))
    #model.add(BatchNormalization())
    #model.add(Dropout(0.5))
    model.add(Dense(2,activation='softmax'))
    learning_rate = 0.01
    momentum = 0.9
    sgd = SGD(lr=learning_rate,momentum = momentum,nesterov=False)
    model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
    callback = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy',patience=3)
    hist=model.fit(train_images,train_labels1,batch_size,epochs,verbose,callbacks=[callback],validation_data=(X_valid_images,y_valid))
    return model,hist
```

```
model,history = build_model_final(X_train_1,y_train_1,X_validate,y_validate)
```

```
test_loss,test_acc=model.evaluate(X_test,y_test)
print('test_acc:',test_acc)
```

```
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
```

```
with open('model.tflite', 'wb') as f:
    f.write(tflite_model)
```
